```python
# =========================================================
# PERSONAL EXPENSE TRACKER PROJECT
# =========================================================

# -----------------------------
# IMPORT REQUIRED MODULE
# -----------------------------

# #json module is used to save and load data from a file
import json

# -----------------------------
# GLOBAL VARIABLES
# -----------------------------

# #This list will store all expenses
expenses = []

# #Monthly budget variable
monthly_budget = 0
```

```python
# =========================================================
# FUNCTION 1: LOAD DATA FROM FILE
# =========================================================

def load_data():
    """
    #This function loads expense data and budget
    #from a JSON file when the program starts
    """

    global expenses
    global monthly_budget

    try:
        # #Open the JSON file in read mode
        with open("expenses.json", "r") as file:

            # #Load data from file
            data = json.load(file)

            # #Store loaded values
            expenses = data["expenses"]           #expenses stores all expense records
            monthly_budget = data["budget"]       #monthly_budget stores user budget

            print("\nData loaded successfully!")

    except FileNotFoundError:
        # #If file does not exist
        print("\nNo previous data found. Starting fresh.")
```

```python

# =========================================================
# FUNCTION 2: SAVE DATA TO FILE
# =========================================================

def save_data():
    """
    #This function saves expense data and budget
    #into a JSON file
    """

    # #Create dictionary to store all data
    data = {
        "expenses": expenses,
        "budget": monthly_budget
    }

    # #Open file in write mode
    with open("expenses.json", "w") as file:

        # #Write data into JSON file
        json.dump(data, file, indent=4)

    print("\nData saved successfully!")
```

```python
# =========================================================
# FUNCTION 3: ADD EXPENSE
# =========================================================

def add_expense():
    """
    #This function takes expense details from user
    #and stores them into the expenses list
    """

    # #Take date input
    date = input("Enter date (YYYY-MM-DD): ")

    # #Take category input
    category = input("Enter category (Food/Travel/Shopping/etc): ")

    # #Take amount input
    amount = float(input("Enter amount spent: "))

    # #Take description input
    description = input("Enter description: ")

    # #Create dictionary for one expense
    expense = {
        "date": date,
        "category": category,
        "amount": amount,
        "description": description
    }

    # #Add expense into list
    expenses.append(expense)

    print("\nExpense added successfully!")
```

```python
# =========================================================
# FUNCTION 4: VIEW ALL EXPENSES
# =========================================================

def view_expenses():
    """
    #This function displays all saved expenses
    """

    # #Check if list is empty
    if len(expenses) == 0:
        print("\nNo expenses found!")
        return

    print("\n------ ALL EXPENSES ------")

    # #Loop through all expenses
    for expense in expenses:

        print(f"""
Date       : {expense['date']}
Category   : {expense['category']}
Amount     : ₹{expense['amount']}
Description: {expense['description']}
------------------------------
""")
```

```python

# =========================================================
# FUNCTION 5: SET MONTHLY BUDGET
# =========================================================

def set_budget():
    """
    #This function allows user to set monthly budget
    """

    global monthly_budget

    # #Take budget input
    monthly_budget = float(input("Enter monthly budget: ₹"))

    print("\nBudget updated successfully!")


```

```python

# =========================================================
# FUNCTION 6: VIEW SPENDING SUMMARY
# =========================================================

def spending_summary():
    """
    #This function calculates total spending
    #and compares it with budget
    """

    # #Calculate total expense amount
    total_spent = 0

    # #Loop through expenses
    for expense in expenses:
        total_spent += expense["amount"]

    print("\n------ SPENDING SUMMARY ------")

    print(f"Total Spent   : ₹{total_spent}")
    print(f"Monthly Budget: ₹{monthly_budget}")

    # #Calculate remaining balance
    remaining = monthly_budget - total_spent

    print(f"Remaining     : ₹{remaining}")

    # #Budget warning
    if total_spent > monthly_budget:
        print("WARNING: You exceeded your budget!")
    else:
        print("Good job! You are within budget.")
```

```python
# =========================================================
# FUNCTION 7: CATEGORY-WISE EXPENSE REPORT
# =========================================================

def category_report():
    """
    #This function shows total spending category-wise
    """

    # #Dictionary to store category totals
    category_totals = {}

    # #Loop through all expenses
    for expense in expenses:

        category = expense["category"]
        amount = expense["amount"]

        # #If category already exists
        if category in category_totals:
            category_totals[category] += amount

        else:
            category_totals[category] = amount

    print("\n------ CATEGORY REPORT ------")

    # #Display category totals
    for category, total in category_totals.items():
        print(f"{category} : ₹{total}")
```

```python
# =========================================================
# MAIN MENU FUNCTION
# =========================================================

def menu():
    """
    #This function displays menu continuously
    """

    # #Load old data at start
    load_data()

    while True:

        print("""
===============================
 PERSONAL EXPENSE TRACKER
===============================
1. Add Expense
2. View Expenses
3. Set Monthly Budget
4. View Spending Summary
5. Category-wise Report
6. Save Data
7. Exit
===============================
""")

        # #Take user choice
        choice = input("Enter your choice: ")

         # -----------------------------
        # CHECK USER CHOICE
        # -----------------------------

        if choice == "1":
            add_expense()

        elif choice == "2":
            view_expenses()

        elif choice == "3":
            set_budget()

        elif choice == "4":
            spending_summary()

        elif choice == "5":
            category_report()

        elif choice == "6":
            save_data()

        elif choice == "7":

            # #Save before exiting
            save_data()

            print("\nThank you for using Expense Tracker!")
            break

        else:
            print("\nInvalid choice! Please try again.")


# =========================================================
# PROGRAM STARTS HERE
# =========================================================

menu()
```

## Code using Text File

```python
# =========================================================
# PERSONAL EXPENSE TRACKER (WITHOUT JSON)
# =========================================================

# #List to store all expenses
expenses = []

# #Variable to store monthly budget
monthly_budget = 0


# =========================================================
# FUNCTION 1: LOAD DATA FROM FILE
# =========================================================

def load_data():

    """
    #This function reads expense data
    #from a text file
    """

    global expenses

    try:

        # #Open file in read mode
        with open("expenses.txt", "r") as file:

            # #Read all lines from file
            lines = file.readlines()

            # #Loop through each line
            for line in lines:

                # #Remove newline character
                line = line.strip()

                # #Split line using comma
                data = line.split(",")

                # #Create dictionary
                expense = {
                    "date": data[0],
                    "category": data[1],
                    "amount": float(data[2]),
                    "description": data[3]
                }

                # #Add expense to list
                expenses.append(expense)

        print("\nData loaded successfully!")

    except FileNotFoundError:

        # #If file not found
        print("\nNo previous data found.")
```

```python
# =========================================================
# FUNCTION 2: SAVE DATA TO FILE
# =========================================================

def save_data():

    """
    #This function saves all expenses
    #into a text file
    """

    # #Open file in write mode
    with open("expenses.txt", "w") as file:

        # #Loop through expenses
        for expense in expenses:

            # #Convert dictionary into string line
            line = f"{expense['date']},{expense['category']},{expense['amount']},{expense['description']}\n"

            # #Write line into file
            file.write(line)

    print("\nData saved successfully!")
```

```python

# =========================================================
# FUNCTION 3: ADD EXPENSE
# =========================================================

def add_expense():

    """
    #This function takes expense details
    #from the user
    """

    # #Input date
    date = input("Enter date (YYYY-MM-DD): ")

    # #Input category
    category = input("Enter category: ")

    # #Input amount
    amount = float(input("Enter amount spent: "))

    # #Input description
    description = input("Enter description: ")

    # #Create expense dictionary
    expense = {
        "date": date,
        "category": category,
        "amount": amount,
        "description": description
    }

    # #Add expense to list
    expenses.append(expense)

    print("\nExpense added successfully!")
```

```python

# =========================================================
# FUNCTION 4: VIEW EXPENSES
# =========================================================

def view_expenses():

    """
    #This function displays all expenses
    """

    # #Check if expenses list is empty
    if len(expenses) == 0:
        print("\nNo expenses found.")
        return

    print("\n------ ALL EXPENSES ------")

    # #Loop through expenses
    for expense in expenses:

        print(f"""
Date       : {expense['date']}
Category   : {expense['category']}
Amount     : ₹{expense['amount']}
Description: {expense['description']}
------------------------------
""")
```

```python
# =========================================================
# FUNCTION 5: SET MONTHLY BUDGET
# =========================================================

def set_budget():

    """
    #This function sets monthly budget
    """

    global monthly_budget

    # #Take budget input
    monthly_budget = float(input("Enter monthly budget: ₹"))

    print("\nBudget set successfully!")
```

```python

# =========================================================
# FUNCTION 6: SPENDING SUMMARY
# =========================================================

def spending_summary():

    """
    #This function calculates total expenses
    """

    total_spent = 0

    # #Loop through expenses
    for expense in expenses:

        total_spent += expense["amount"]

    print("\n------ SPENDING SUMMARY ------")

    print(f"Total Spent   : ₹{total_spent}")
    print(f"Monthly Budget: ₹{monthly_budget}")

    # #Calculate remaining budget
    remaining = monthly_budget - total_spent

    print(f"Remaining     : ₹{remaining}")

    # #Check budget status
    if total_spent > monthly_budget:

        print("WARNING: Budget exceeded!")

    else:

        print("You are within budget.")
```

```python

# =========================================================
# FUNCTION 7: CATEGORY-WISE REPORT
# =========================================================

def category_report():

    """
    #This function shows category-wise spending
    """

    # #Dictionary for totals
    category_totals = {}

    # #Loop through expenses
    for expense in expenses:

        category = expense["category"]
        amount = expense["amount"]

        # #Add amount category-wise
        if category in category_totals:

            category_totals[category] += amount

        else:

            category_totals[category] = amount

    print("\n------ CATEGORY REPORT ------")

    # #Display category totals
    for category, total in category_totals.items():

        print(f"{category} : ₹{total}")

```

```python
# =========================================================
# MAIN MENU FUNCTION
# =========================================================

def menu():

    """
    #Main menu function
    """

    # #Load saved data
    load_data()

    while True:

        print("""
================================
 PERSONAL EXPENSE TRACKER
================================
1. Add Expense
2. View Expenses
3. Set Monthly Budget
4. Spending Summary
5. Category Report
6. Save Data
7. Exit
================================
""")

        # #Take user choice
        choice = input("Enter your choice: ")

        # #Menu conditions

        if choice == "1":

            add_expense()

        elif choice == "2":

            view_expenses()

        elif choice == "3":

            set_budget()

        elif choice == "4":

            spending_summary()

        elif choice == "5":

            category_report()

        elif choice == "6":

            save_data()

        elif choice == "7":

            # #Save before exiting
            save_data()

            print("\nThank you for using Expense Tracker!")

            break

        else:

            print("\nInvalid choice. Try again.")

# =========================================================
# PROGRAM STARTS HERE
# =========================================================

menu()
break()
```
